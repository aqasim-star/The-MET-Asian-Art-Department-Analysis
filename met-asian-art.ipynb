```python
# met api for asian art using both the department name and the department id
```

```python
base_url = 'https://collectionapi.metmuseum.org'
search_path = '/public/collection/v1/search'
object_path = '/public/collection/v1/objects/'
import requests
```

```python
# USING DEPARTMENT ID
```

```python
query = '?q=*&departmentId=6'
departmentid_request = requests.get(f'{base_url}{search_path}{query}')
departmentid_parsed = departmentid_request.json()
departmentid_parsed
```

```python
ids = departmentid_parsed['objectIDs']
```

```python
artworks = []
```

```python
for item in ids:
    object_result = requests.get(f'{base_url}{object_path}{item}')
    result_parsed = object_result.json()
    artworks.append(result_parsed)
```

```python
artworks
```
