```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('met-asian-art.csv')
```

```python
df
```

```python
df.sort_values('culture')
```

```python
df.value_counts('culture').plot(kind='bar')
```

```python
df.value_counts('isPublicDomain').plot(kind='pie')
```

```python
counts = df['isPublicDomain'].value_counts()
percent = (df['isPublicDomain'].value_counts(normalize=True) * 100).round(2)

pd.DataFrame({'count': counts, 'percent': percent})
```

```python
df.value_counts('classification').plot(kind='bar')
```

```python
df.value_counts('classification').nlargest(10).plot(kind='bar')
```
